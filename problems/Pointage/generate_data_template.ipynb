```python
import os
import sys
from random import randint
from random import random
```

```python
has_test_cases = False
amount_of_test_cases = 1000

amount_of_inputs = 50

start_input_index = 10
```

```python
def randDouble(min, max, precision=2):
    return round(min + (max - min) * random(), precision)

def randInt(min, max):
    return randint(min, max)
```

```python
def generate_case():
  test_case = ""
  test_case += str(randDouble(-1000, 1000)) + " " + str(randDouble(-1000, 1000)) + "\n" # x1 y1
  test_case += str(randDouble(-1000, 1000)) + " " + str(randDouble(-1000, 1000)) + "\n" # x2 y2
  test_case += str(randDouble(-1000, 1000)) + "\n" # x3 => y3
  return test_case

def generate_input():
  
  for i in range(0, amount_of_inputs):
    input_text = ""

    if has_test_cases:
      # number of test cases
      input_text += str(amount_of_test_cases) + "\n"
      # the t test cases
      for j in range(0, amount_of_test_cases):
        input_text += generate_case()
      
    else:
      input_text += generate_case() 
    
    print(input_text)    
    filename = "input/input" + str(i+start_input_index) + ".txt"

    with open(filename, "w") as file:
      file.write(input_text)
```

```python
generate_input()
```
