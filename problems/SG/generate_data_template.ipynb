```python
import os
import sys
from random import randint
from random import random
```

```python
has_test_cases = False
amount_of_test_cases = 10

amount_of_inputs = 20

start_input_index = 130
```

```python
def randDouble(min, max, precision=2):
    return round(min + (max - min) * random(), precision)

def randInt(min, max):
    return randint(min, max)
```

```python
def generate_case():
  test_case = ""
  # generate the case
  # n, m, p : the number of students, machines, and teachers
  # m number : machine capacities
  # n number : teacher capacities
  # followed by p lines of teacher machines
  # ith line starts with the number of machines the teacher i has, followed by the machines
  
  n = randInt(1e6, 1e9)
  m = randInt(400, 500)
  p = randInt(400, 500)

  test_case += str(n) + " " + str(m) + " " + str(p) + "\n"

  machine_capacities = [randInt(50, 100) for i in range(m)]
  test_case += " ".join(str(x) for x in machine_capacities) + "\n"

  teacher_capacities = [randInt(50, 100) for i in range(p)]
  test_case += " ".join(str(x) for x in teacher_capacities) + "\n"

  for i in range(p):
    teacher_machines = [randInt(0, 1)*randInt(0,1)*randInt(0,1) for i in range(randInt(1, m))]
    machines = []
    for j in range(len(teacher_machines)):
      if teacher_machines[j] == 1:
        machines.append(j)
    test_case += str(len(machines)) + " " + " ".join(str(x) for x in machines) + "\n"

  return test_case

def generate_input():
  
  for i in range(0, amount_of_inputs):
    input_text = ""

    if has_test_cases:
      # number of test cases
      input_text += str(amount_of_test_cases) + "\n"
      # the t test cases
      for j in range(0, amount_of_test_cases):
        input_text += generate_case()
      
    else:
      input_text += generate_case() 
    
    print(input_text)    
    filename = "input/input" + str(i+start_input_index) + ".txt"

    with open(filename, "w") as file:
      file.write(input_text)
```

```python
generate_input()
```
